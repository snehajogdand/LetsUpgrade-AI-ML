# Assignment 1

Logistic Regression on Bank Personal Loan

```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_excel('Bank_Personal_Loan_Modelling.xlsx',sheet_name='Data')
```

```python
dataset.drop(columns='ID')
dataset.drop(columns='ZIP Code')
```

```python
y=dataset['Personal Loan']
```

```python
x=dataset[['Age','Experience','Income','Family','CCAvg','Education','Mortgage','Securities Account','CD Account','Online',
           'CreditCard']]
```

```python
import statsmodels.api as sm
```

```python
x1=sm.add_constant(x)
```

```python
Logistic=sm.Logit(y,x1)
```

```python
result=Logistic.fit()
```

```python
result.summary()
```

# Assignment 3

Linear Regression on Real Estate Analysis

```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_excel('Linear Regression.xlsx',sheet_name=0)
```

```python
dataset
```

```python
y=dataset['price']
```

```python
x=dataset[['sqft_living','bedrooms','bathrooms','floors']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,random_state=True,train_size=0.1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mul_reg=LinearRegression()
```

```python
mul_reg.fit(x_train,y_train)
```

```python
y_pred=mul_reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2=r2_score(y_test,y_pred)
```

```python
RMSE=np.sqrt(mean_squared_error(y_test,y_pred))
```
